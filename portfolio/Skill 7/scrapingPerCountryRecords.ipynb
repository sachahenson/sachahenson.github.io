```python
import requests, pandas as pd, numpy as np
```

```python
tables=pd.read_html('https://www.worldathletics.org/records/by-category/world-records')
```

```python
## assigning dataframes to each table
one=tables[0]
two=tables[1]
three=tables[2]
four=tables[3]
five=tables[4]
```

```python
## putting dataframes into one df
all=one.append(two).append(three).append(four).append(five)
all
```

```python
## dropping irrelevant columns
all.drop(columns=['Discipline','Progression','Perf','Wind','Competitor','DOB','Venue','Date'], inplace=True)
```

```python
all
```

```python
## counting per country records
all['Country'].value_counts()
```

```python
## adding count of total records to df
all['Total records'] = all.groupby('Country')['Country'].transform('count')
all
```

```python
## dropping duplicates of country value
all.drop_duplicates(['Country'], inplace=True)
```

```python
## descending order of total records
all.sort_values(by=['Total records'], ascending=False, inplace=True)
```

```python
  ## resetting index
  all.reset_index(inplace=True)
```

```python
all
```

```python
## dropping columns created from messing around with index
all.drop(columns=['index'])
```

```python
all
```

```python
## save
all.to_csv('percountry_records.csv')
```
