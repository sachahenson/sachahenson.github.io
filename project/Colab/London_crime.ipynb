#Plotting and best fit of annual crime by borough and house prices

```python
## Importing relevant packages
from sklearn.pipeline import make_pipeline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)
from scipy.stats import linregress
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
```

```python
## LOGGED HOUSE PRICES X CRIME
## Setting variables: crime total and logged house prices
x = crime_houseprices_annual['Frequency']
y = crime_houseprices_annual['log_value']

## Best fit line of crime total and logged house prices
model.fit(x[:, np.newaxis], y)

xfit = np.linspace(9000, 70000)
yfit = model.predict(xfit[:, np.newaxis])

plt.scatter(x, y)
plt.plot(xfit, yfit);
```

```python
print("Model slope:    ", model.coef_[0])
print("Model intercept:", model.intercept_)
```
